```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

def preprocessing_txt(txt):
    txt = txt.lower()
    txt = re.sub(r'<.*?>', ' ', txt)
    txt = re.sub(r'[^a-zA-Z]', ' ', txt)
    txt = txt.strip()
    
    
    txt = word_tokenize(txt)
    
    
    txt = [word for word in txt if word not in stop_words]
    stemming=PorterStemmer()
    txt =[stemming.stem(word) for word in txt]
  
    
    return ' '.join(txt)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_df=0.95,max_features=5000,ngram_range=(1,3))
word_count_vectors=cv.fit_transform(docs)
```

```python
from sklearn.feature_extraction.text import TfidfTransformer
tftidf_transformer =TfidfTransformer(smooth_idf=True,use_idf=True)
tfidf_transformer  =tftidf_transformer.fit(word_count_vectors)
```

```python
cv.get_feature_names_out()
```

```python
# get feature names
feature_names = cv.get_feature_names_out()


def get_keywords(idx, docs, topN=10):
    # transform document to tf-idf
    doc_tfidf = tfidf_transformer.transform(
        cv.transform([docs[idx]])
    )

    # convert sparse matrix to COO format
    doc_tfidf = doc_tfidf.tocoo()

    # pair feature index with tf-idf score
    tuples = zip(doc_tfidf.col, doc_tfidf.data)

    # sort by score (descending)
    sorted_items = sorted(tuples, key=lambda x: x[1], reverse=True)

    # take top N keywords
    sorted_items = sorted_items[:topN]

    score_vals = []
    feature_vals = []

    for idx, score in sorted_items:
        score_vals.append(round(score, 3))
        feature_vals.append(feature_names[idx])

    # build result dictionary
    results = {}
    for i in range(len(feature_vals)):
        results[feature_vals[i]] = score_vals[i]

    return results


def print_keywords(idx, keywords, df):
    print("\n======= TITLE =======")
    print(df['title'][idx])

    print("\n======= ABSTRACT =======")
    print(df['abstract'][idx])

    print("\n======= KEYWORDS =======")
    for k, v in keywords.items():
        print(f"{k} : {v}")


# run
idx = 4995
keywords = get_keywords(idx, docs)
print_keywords(idx, keywords, df)
```
